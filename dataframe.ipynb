## DataFrame

1. Create DF using List

```python
import pandas as pd
lst = ['kavya', 'is', 'learning', 'sybau']
lst = ['kavya', 'is', 'learning', 'sybau']
df = pd.DataFrame(lst)
print(df)
```

2. Create DF from dict of ndarray/lists

```python
import pandas as pd
data = {'Name':['Tom', 'nick', 'krish', 'jack'], 'Age':[20,21,19,18]}
df = pd.DataFrame(data)
print(df)
```

3. Column Selection

```python
import pandas as pd
 
data = {'Name':['Jai', 'Princi', 'Gaurav', 'Anuj'],
        'Age':[27, 24, 22, 32],
        'Address':['Delhi', 'Kanpur', 'Allahabad', 'Kannauj'],
        'Qualification':['Msc', 'MA', 'MCA', 'Phd']}
 
df = pd.DataFrame(data)
 
print(df[['Name', 'Qualification']])
```

4. Row Selection

```python
import pandas as pd
data = pd.read_csv("nba.csv", index_col="Name")
first = data.loc["Avery Bradley"]
second = data.loc["R.J. Hunter"]
print(first, "\n\n", second)
```

5. Indexing using indexing operator to access columns - []

```python
import pandas as pd
data = pd.read_csv("nba.csv", index_col="Name")
first = data["Age"]
print(first)
```

6. Indexing by label - .loc[]

```python
import pandas as pd
data = pd.read_csv("nba.csv", index_col="Name")
first = data.loc["Avery Bradley"]
second = data.loc["R.J. Hunter"]
print(first, "\n\n", second)
```

7. Indexing by integer position - .iloc[]

```python
import pandas as pd
data = pd.read_csv("nba.csv", index_col="Name")
row2 = data.iloc[3]
print(row2)
```

8. Check for missing values - 
isnull() - returns True for NaN, False otherwise
notnull() - returns False for NaN, True otherwise

```python
import numpy as np
import pandas as pd
dict = {'First Score':[100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score':[np.nan, 40, 80, 98]}
df = pd.DataFrame(dict)
print(df.isnull())
```

9. Fill missing values with fillna(), replace(), interpolate()

```python
import numpy as np
import pandas as pd
data = pd.read_csv("nba.csv", index_col="Name")
dict = {'First Score':[100, 90, np.nan, 95],
        'Second Score': [30, 45, 56, np.nan],
        'Third Score':[np.nan, 40, 80, 98]}
df = pd.DataFrame(dict)
print(df.fillna(0))
```
